```python
popl_data = pd.read_csv('./dataset/countries.csv')
popl_data
```

```python
us = popl_data[popl_data.country == 'United States']
china = popl_data[popl_data.country == 'China']
```

```python
plt.plot(us.year, us.population/10**6)
plt.plot(china.year, china.population/10**6)
plt.title('U.S vs China Population graph from year 1952-2007')
plt.xlabel('Year')
plt.ylabel('population in millions (10^6)')
plt.legend(['this is U.S','this is China'])
plt.show()
```

```python
#percentage growth
(us.population.iloc[10]/us.population.iloc[0])*100 - 100
```

```python
plt.plot(us.year, (us.population/us.population.iloc[0])*100 - 100)
plt.plot(china.year, (china.population/china.population.iloc[0])*100 - 100)
plt.title('U.S vs China Population growth rate from year 1952-2007')
plt.xlabel('Year')
plt.ylabel('population growth in %')
plt.legend(['this is U.S growth curve','this is China growth curve'])
plt.show()
```
